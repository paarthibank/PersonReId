```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPooling2D,Concatenate
from keras.optimizers import adam_v2
import cv2
import os
import pandas as pd
from keras.preprocessing.image import ImageDataGenerator
from keras.applications.mobilenet import MobileNet
from tensorflow.keras import Input, Model

from tensorflow.keras.utils import to_categorical
```

```python

in1 = Input(shape=(128,64,3))
in2 = Input(shape=(128,64,3))
#     mobilenet= MobileNet(include_top=False,
#                         input_shape=(128,256,3))
mc1=Conv2D(1,kernel_size=(3,3))(in1)
mc2=Conv2D(1,kernel_size=(3,3))(in2)
m1=MaxPooling2D()(mc1)
m2=MaxPooling2D()(mc2)
f1=Flatten()(m1)
f2=Flatten()(m2)
co= Concatenate()([f1, f2])
out=Dense(2, activation='softmax')(co)

model = Model(inputs=[in1,in2], outputs=out)
model.summary()

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    
i=0
for filename1 in os.listdir('./Market-1501-v15.09.15/bounding_box_train'):
    imgs1=[]
    imgs2=[]
    y=[]
    img1 = cv2.imread(os.path.join(sourceFolder,filename1))
    i+=1
    if i>=100:
        break
    print(i)
    if img1 is not None:
        for filename2 in os.listdir(sourceFolder):
            img2 = cv2.imread(os.path.join(sourceFolder,filename2))
            if img2 is not None and filename1!=filename2:
                imgs1.append(img1)
                imgs2.append(img2)
                if filename1[0]==filename2[0] and filename1[1]==filename2[1] and filename1[2]==filename2[2] and filename1[3]==filename2[3]:
                    y.append(1)
                else:
                    y.append(0)
    imgs1=np.array(imgs1)
    imgs2=np.array(imgs2)
    y=pd.DataFrame(y)
    y = to_categorical(y)
    hist = model.fit([imgs1,imgs2],y, validation_split=0.2, epochs=20,)
    hist = hist.history
    plt.plot(hist["accuracy"])
    plt.plot(hist["val_accuracy"])
    plt.title("Accuracy plot")
    plt.legend(["train", "test"])
    plt.xlabel("epoch")
    plt.ylabel("accuracy")
    plt.show()
                
```
